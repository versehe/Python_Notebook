```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic

```python
# put multiple graphs in single figure
fig = plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
# plot =defaults on the last fig
plt.plot(np.arange(10))
# to plot on specific fig
ax1.scatter(np.arange(30),np.arange(30)*2)
# set ticks
ax1.set_xticks([0,5,10,15,20,25,30,35])
ax1.set_yticklabels(['Zero','Two','Four','Superise','Ever'])
ax1.set_title('Example')
ax1.set_ylabel('No Sense')

# plot multiple lines in one graph
ax2.plot(np.arange(10),label = 'First')
ax2.plot(np.arange(15),label = 'Second')
ax2.plot(np.arange(6),label = 'Third')
ax2.legend(loc='best')

plt.show()
#plt.savefig('example.png')
```

```python
# plot a dataframe/series
purchase_1 = pd.Series({'Name': 'Chris', 'Cost': 22.50})
purchase_2 = pd.Series({'Name': 'Kevyn','Cost': 2.50})
purchase_3 = pd.Series({'Name': 'Vinod', 'Cost': 5.00})

df = pd.DataFrame([purchase_1, purchase_2, purchase_3], index=['Store 1', 'Store 1', 'Store 2'])
df.plot()
plt.show()
```

```python
# create a grid for plots
fig,axes = plt.subplots(2,3)
plt.show()
```

## Line Chart

```python
# simple line chart
plt.plot(np.arange(10),'ko--')
# 'ko--' defines a black dash line with points

plt.show()
```

## Bar Chart

```python
fig,axes = plt.subplots(2,1)
data = pd.Series(np.random.rand(16))
data.plot(kind='bar', ax =axes[0])
data.plot(kind='barh', ax =axes[1])
plt.show()
```

```python
df = pd.DataFrame(np.random.rand(6,2),index=[1,2,3,4,5,6],
                 columns = pd.Index(['A','B'], name= 'Genius' ))
df.plot(kind='bar')
plt.show()
```
