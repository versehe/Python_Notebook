```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic

```python
# plot a dataframe/series
purchase_1 = pd.Series({'Name': 'Chris', 'Cost': 22.50})
purchase_2 = pd.Series({'Name': 'Kevyn','Cost': 2.50})
purchase_3 = pd.Series({'Name': 'Vinod', 'Cost': 5.00})

df = pd.DataFrame([purchase_1, purchase_2, purchase_3], index=['Store 1', 'Store 1', 'Store 2'])
df.plot()
plt.show()
```

```python
# create a grid for plots
fig,axes = plt.subplots(2,3)
ax = plt.subplot(2, 3, 4)
ax.text(0.5, 0.5, str((2, 3,4)),
           fontsize=18, ha='center') 
plt.show()
```

## Line Chart

```python
# simple line chart
plt.plot(np.arange(10), color='blue', linestyle='dashdot', label='x')
plt.plot(np.arange(10), np.cos(np.arange(10)), label='cos(x)')
# set Axes Limits
plt.xlim(-1, 11)
plt.ylim(-1.5, 11.5)
#or: plt.axis([-1, 11, -1.5, 11.5])

plt.xlabel("No Sense")
plt.ylabel("cos(x)")
plt.title('Example')
plt.legend() # need to mark label to show legends

plt.show()
```

## Bar Chart

```python
fig,axes = plt.subplots(2,1)
data = pd.Series(np.random.rand(16))
data.plot(kind='bar', ax =axes[0])
data.plot(kind='barh', ax =axes[1])
plt.show()
```

```python
df = pd.DataFrame(np.random.rand(6,2),index=[1,2,3,4,5,6],
                 columns = pd.Index(['A','B'], name= 'Genius' ))
df.plot(kind='bar')
plt.show()
```

## Scatter

```python
plt.plot(np.arange(10), np.sin(np.arange(10)), 'o', color='black',
         markersize=15, markerfacecolor='white', markeredgecolor='black', markeredgewidth=2) #set marker format
plt.ylim(-1.2, 1.2)
plt.show()
```

```python
# plot with scatter function is more powerful
# it can control the properties of each individual point (size, color, etc.)

rng = np.random.RandomState(1230) # set random seed
x = rng.randn(100)  #Return samples from standard normal distribution
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,cmap='viridis') #alpha is transparency level
# viridis is the default colormap
plt.colorbar()  # show color scale

plt.show()
```

## Pie Chart

```python
plt.pie([15, 30, 45, 10],  labels=['Frogs', 'Hogs', 'Dogs', 'Logs'], autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.axis('equal')  # Equal ensures that pie is drawn as a circle
plt.legend()

plt.show()
```

## Error Bar

```python
x = np.linspace(0, 10, 5)  #start = 0, stop = 10, number = 5
y = np.sin(x) + 0.8* np.random.randn(5)

plt.errorbar(x, y, yerr=0.8, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0)

plt.show()
```

## Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, alpha=0.5)

plt.show()
```

```python
# compare multiple distributions in one plot
plt.hist(np.random.normal(0, 0.8, 1000), histtype='stepfilled', alpha=0.3, normed=True, bins=40, label='DistA')
plt.hist(np.random.normal(-2, 1, 1000), histtype='stepfilled', alpha=0.3, normed=True, bins=40, label='DistB')
plt.hist(np.random.normal(3, 2, 1000), histtype='stepfilled', alpha=0.3, normed=True, bins=40, label='DistC')

plt.legend(loc='upper left', frameon=False, ncol=2)
plt.show()
```

## ax Method : multiple graphs

```python
# put multiple graphs in single figure
fig = plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
# plot defaults on the last fig
plt.plot(np.arange(10))
# to plot on specific fig
ax1.scatter(np.arange(30),np.arange(30)*2)

# most functions can be called as plt method, but not set limits, labels, and titles.
ax1.set_title('Example')
ax1.set_ylabel('No Sense')
ax1.set_xlim(0,40)
# or: ax1.set_xticks([0,5,10,15,20,25,30,35,40])
ax1.set_yticklabels(['Zero','Two','Four','Superise','Ever'])

# plot multiple lines in one graph
ax2.plot(np.arange(10),label = 'First')
ax2.plot(np.arange(15),label = 'Second')
ax2.plot(np.arange(6),label = 'Third')
ax2.legend(loc='best')

plt.show()
#plt.savefig('example.png')
```
